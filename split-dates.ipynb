```python
import pandas as pd
```

```python
df = pd.read_csv('editors-output-2018-09-09.csv')
```

```python
df.columns
```

```python
df2 = pd.DataFrame(df['Authorized Name'].apply(lambda x: str(x).split(', ')))
```

```python
df2
```

```python
df3 = pd.DataFrame(df2['Authorized Name'].values.tolist(),columns=['name1','name2','date'])
```

```python
df3.columns
```

```python
df3
```

```python
df4 = pd.DataFrame(df3['date'].apply(lambda x: str(x).split('-')),columns=['date'])
```

```python
df4
```

```python
df4 = pd.DataFrame(df4['date'].values.tolist(),columns=['bdate','ddate'])
```

```python
df4
```

```python
df5 = pd.concat([df3,df4], axis=1, sort=False)
```

```python
df5
```

```python
df5.to_csv('editors-dates-output-2018-09-09.csv')
```
